```python
# Task 2: Mark the success/failed launches for each site on the map

# Task 2: Mark the success/failed launches for each site on the map

# Group the launch data by launch site and outcome (success or failure)
launch_outcome_counts = spacex_df.groupby(['Launch Site', 'class']).size().reset_index(name='Counts')

# Pivot the data to have success and failure counts as separate columns
launch_outcome_counts_pivot = launch_outcome_counts.pivot(index='Launch Site', columns='class', values='Counts').fillna(0)

# Calculate total launches for each site
launch_outcome_counts_pivot['Total'] = launch_outcome_counts_pivot.sum(axis=1)

# Initialize the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Define colors for markers (green for success, red for failure)
color_dict = {0: 'red', 1: 'green'}

# Add markers for each launch site
for index, row in launch_sites_df.iterrows():
    site = row['Launch Site']
    lat = row['Lat']
    long = row['Long']
    success_count = int(launch_outcome_counts_pivot.loc[site, 1])
    failure_count = int(launch_outcome_counts_pivot.loc[site, 0])
    total_count = int(launch_outcome_counts_pivot.loc[site, 'Total'])
    html = f'<b>{site}</b><br>Total Launches: {total_count}<br>Successes: {success_count}<br>Failures: {failure_count}'
    folium.Marker(
        location=[lat, long],
        popup=html,
        icon=folium.Icon(color=color_dict[1], icon='info-sign')
    ).add_to(site_map)

# Display the map
site_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python

# Initialize the map
# site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Add markers for each launch record to the MarkerCluster
for index, row in spacex_df.iterrows():
    site = row['Launch Site']
    lat = row['Lat']
    long = row['Long']
    launch_outcome = row['class']
    
    # Determine marker color based on launch outcome
    if launch_outcome == 1:
        color = 'green'  # Successful launch
    else:
        color = 'red'  # Failed launch
    
    # Create HTML content for marker popup
    html = f'<b>{site}</b><br>Outcome: {"Success" if launch_outcome == 1 else "Failure"}'
    
    # Add marker to the MarkerCluster
    folium.Marker(
        location=[lat, long],
        popup=html,
        icon=folium.Icon(color=color, icon='info-sign')
    ).add_to(marker_cluster)

# Add MarkerCluster to the map
marker_cluster.add_to(site_map)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities

from math import radians, sin, cos, sqrt, atan2

def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = 6371 * c  # Radius of the Earth in kilometers
    return distance

# Example coordinates for launch site and proximity
launch_site_lat = 28.573255
launch_site_lon = -80.646895
proximity_lat = 28.562302
proximity_lon = -80.577356

# Calculate distance between launch site and proximity
distance = calculate_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)
print("Distance between launch site and proximity:", distance, "km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
